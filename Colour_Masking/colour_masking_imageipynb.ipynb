```python


img = cv2.imread("/content/gdrive/MyDrive/arwaa img/hs.jpg")

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_bound = np.array([5, 100, 100])
upper_bound = np.array([15, 255, 255])

mask = cv2.inRange(hsv, lower_bound, upper_bound)

kernel = np.ones((5, 5), np.uint8)
mask = cv2.erode(mask, kernel, iterations=1)

result = cv2.bitwise_and(img, img, mask=mask)

plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title("Image")
plt.subplot(132), plt.imshow(mask, cmap='gray'), plt.title('Mask')
plt.subplot(133), plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB)), plt.title('Result')
plt.show()
```
